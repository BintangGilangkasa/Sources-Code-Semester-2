```python
#Mengimport library
import mysql.connector
from tabulate import tabulate

#Membuat connecting to database
db_connection = mysql.connector.connect (
    host = "localhost",
    user = "root",
    password = "",
    database = "praktek_mandiri"
)

# Membuat cursor
cursor = db_connection.cursor()

# Menulis dan Mengeksekusi Query
query = "SELECT * FROM anggota WHERE no_hp is null"
cursor.execute(query)

# Mengambil semua baris pada query
data = cursor.fetchall()
colomns = cursor.column_names

cursor.close()
db_connection.close()

print(tabulate(data, headers=colomns, tablefmt="grid"))
```
